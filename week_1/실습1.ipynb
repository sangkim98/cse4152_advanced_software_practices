# 코드 예시) Random 함수 


```python
import random
import numpy as np
```

```python
#random.random() -> 0.0에서 1.0사이의 실수 중에서 난수값 리턴
print(random.random()) 
# ex) 0.2664520652284096

#random.uniform(a, b) -> 괄호 안 두 수 사이의 실수 중에서 난수값을 리턴
print(random.uniform(10, 30)) 
# ex) 28.038707445386397

#random.randint(a, b) -> 괄호 안 두 수 사이의 정수 중에서 난수값을 리턴
print(random.randint(100,200))
# ex) 106

#random.choice(sample) 함수 -> sample에서 무작위로 하나를 선택하여 리턴
data = [1, 2, 3, 4, 5, 6, 7]
print(random.choice(data))
# ex) 5

#random.sample(sample, n) -> 입력으로 받은 sample 에서 정한 개수만큼  무작위로 뽑아 리턴함
data = [1, 2, 3, 4, 'apple', 'banna']
print(random.sample(data, 3))
# ex) ['apple', 2, 3]
print(random.sample(data, 3))
# ex) ['apple', 3, 'banna']
print(random.sample(data, 5))
# ex) [2, 3, 1, 'apple', 4]
```

# 실습 2. Linear Congruential Generator(LCG) 구현  

Linear Congruential Generator(LCG) 구현  

X_(n+1) = (aX_n+C) mod m

LCG는 다음과 같은 인자들로 유일하게 결정됨  
  - 0 < m, 0 < a < m, 0 <= c < m, 0<=초기값 < m  

최대 주기 m을 가지기 위한 필요충분조건  

  - c와 m이 서로소, a-1이 m의 모든 소인수로 나뉨 
  - m이 4의 배수면 a-1도 4의 배수

```python
def lcg(seed, loop, m=2**32, a=214013, c=2531011):
    for i in range(loop):
        # write answer
        yield seed
        seed = (a*seed + c) % m
```

```python
num_iterations = 100
random_generator = lcg(seed=13, loop=num_iterations, m=2**32)
random_integers = [i for i in random_generator]

fig = plt.figure()
plt.hist(random_integers)
plt.title(f"Check Uniform Distribution of {num_iterations} iterations")
plt.xlabel("Numbers")
plt.ylabel("N of each number")
plt.xlim([0, 2**32])
plt.show()
```

# 실습 4. 주사위 게임

```python
# Inputs
num_simulations = 10000
max_num_rolls = 1000

win_probability = []

data = np.zeros((num_simulations,max_num_rolls+1), dtype=np.intc)

for row in data:
  row[0] = 1000

for person_idx, person in enumerate(data):
  for trial_idx, trial in enumerate(person[:-1]):
    num_wins = 0

    a = random.randint(1,6)
    b = random.randint(1,6)
    
    if a == b:
      data[person_idx][trial_idx+1] = trial + 4
      num_wins += 1
    else:
      data[person_idx][trial_idx+1] = trial - 1

    win_probability.append(num_wins/max_num_rolls)
```

```python
plt.figure(figsize=(12,6))
plt.plot(data.T)

plt.title("Monte Carlo Dice Game [" + str(num_simulations) + " players simulations]")
plt.xlabel("Roll Number")
plt.ylabel("Balance [$]")
plt.xlim([0, max_num_rolls])
```

```python
overall_win_probability = round((sum(win_probability)/num_simulations), 6)
# overall_end_balance = round((sum(data[:][max_num_rolls+1])/max_num_rolls), 2)
overall_end_balance = round(np.mean(data[:][max_num_rolls+1]), 2)

print(f"10000명 플레이어의 {max_num_rolls}번 수행 횟수 시 평균 승률" + "  :  " + colored(f"{overall_win_probability:.2f}","red"))
print(f"10000명 플레이어의 {max_num_rolls}번 수행 횟수 시 평균 잔액 "  + f" : {colored('$','red')}" + colored(f"{overall_end_balance:.2f}","red"))
```

# 실습 6. 계약 손익 예측

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 시뮬레이션 횟수
N_ROUNDS = 10000
results = []
for rnd in range(N_ROUNDS):
    ## write answer
    base_sales = np.random.triangular(1, 3, 9, 1)[0]
    wins = np.random.rand() < 0.5

    increase_in_sales = 0

    if(wins is True):
      increase_in_sales = np.random.triangular(0.25, 0.5, 0.75, 1)[0]

    sales = base_sales + base_sales*increase_in_sales
    results.append(sales)
```

```python
# 모든 시나리오에서 나온 신작 판매량의 평균 
average_sales = np.mean(results)

binned_results = pd.cut(results, bins=[0, 6, 20], labels=[
    "Probability : less than or equal to 600 million won",
    "Probability : greater than 600 million won"])

# 결과물 시각화
sns.barplot(x=binned_results.value_counts().index,
            y=binned_results.value_counts().values)
```

```python
C1 , C2 = binned_results.value_counts()
P1 = round(C1 / N_ROUNDS * 100,1)
P2 = round(C2 / N_ROUNDS * 100,1)

print("6억원 이하의 매출을 얻을 확률 :", colored(f"{P1}","red") )
print("6억원 초과의 매출을 얻을 확률 :", colored(f"{P2}","blue"))
```

# 실습 7 헬스장 수익 예측


```python
import numpy as np
import seaborn as sns
from termcolor import colored
```

```python
np.random.seed(1)
monthly_membership = 50000

N_ROUNDS = 10000
subscription_record = np.zeros((N_ROUNDS, 61))
results = []

for rnd in range(N_ROUNDS):
  subscription_record[rnd][0] = 2000

for round_idx in range(N_ROUNDS):
  accumulated_members = 0
  subscription_data = {"Month": range(1, 60),
                     "Growth": np.random.normal(0, 0.05, 60)}
  historical_data = subscription_data["Growth"]

  for month in subscription_data['Month']:
    random_idx = np.random.randint(1,60)
    subscription_record[round_idx][month] = round((subscription_record[round_idx][month-1] * (1 + historical_data[month])))

  for idx in range(12):
    accumulated_members += subscription_record[round_idx][idx]

  accumulated_revenue = accumulated_members * monthly_membership

  results.append(accumulated_revenue)
```

# 99%의 신뢰도  
- 10000개의 수행 결과를 낮은 수익부터 순서대로 정렬하고 100번째(100/10000)번째의 수익을 확인

```python
print("하위 1%의 수익 : ", colored(f"{sorted(results)[percentile] / 1e8:.4f} 억","red"))
## write answer
##
lower_revenue = sorted(results)[percentile]
print(f"매년 준비해야할 금액 : ",  colored(f"{(12*9.5*1e7 - lower_revenue)/1e8:.4f} 억","red"))
```

```python
ax = sns.distplot(results, kde=False)
ax.set_xlabel("Average revenue across simulated 12 months")
ax.set_ylabel("# of simulations")
plt.title(f"Average annual revenue across simulations : {round(np.mean(results)/1e8, 2)}")

plt.show()
```
